# FFT with Numpy and Matplotlib #
Studying FFT to solve [this puzzle from reddit.com](http://www.rognemedia.no/Puzzle/1991.html) but photoshop plugins are available but decided to solve it with numpy and matplotlib instead.  Why?  For fun!  If you are interested, you can start [here](http://www.rognemedia.no/Puzzle/).

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Loading data: ##
Easiest way to import the data from png image is to use matplotlib.image.
It returns numpy.ndarray.
<br>
##### Note: #####
If you are loading file with multiple channels (i.e. RGBA) it will return 4 arrays of dimension of image.

```python
mag = mpimg.imread('./magnitude.png')
phase = mpimg.imread('./phase.png')
```

## To get complex number to feed into ifft: ##
$$F = mag * (\cos(phase) + i \cdot \sin(phase))$$
$$F = mag * \cos(phase) + i \cdot mag * \sin(phase)$$
Where $F$ is the data in frequency domain, ready to be fed into ifft.

```python
re = mag * np.cos(phase)
im = mag * np.sin(phase)
```

```python
F = re + (1j * im)
F.shape
```

```python
out = np.fft.ifftn(F)
plt.imshow(np.log(np.abs(out)), cmap='Greys')
```

## Quick studies of FT on different images ##
Sample images are from [Introduction to Fourier Transforms for Image Processing](http://www.cs.unm.edu/~brayer/vision/fourier.html).

Use np.log function to correct the scaling.

```python
testImg = mpimg.imread('./cosines.png')
testfft = np.fft.fftn(testImg[:,:,0])


plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(testImg)
plt.subplot(1,2,2)
plt.imshow(np.log(np.abs(np.fft.fftshift(testfft))), cmap='Greys')
```

```python
testImg2 = mpimg.imread('./cosines_grey.png')
testfft2 = np.fft.fftn(testImg2)

plt.figure(1)
plt.subplot(121)
plt.imshow(testImg2, cmap='Greys')
plt.subplot(122)
plt.imshow(np.log(np.abs(np.fft.fftshift(testfft))), cmap='Greys')
```

```python
tryInvert = mpimg.imread('./invert.png')
invfft = np.fft.ifft2(tryInvert[:,:,0])

plt.figure(1)
plt.subplot(121)
plt.imshow(tryInvert)
plt.subplot(122)
plt.imshow(np.log(abs(invfft)), cmap='Greys',  interpolation='nearest')
```
